**Exercice 2** : Ecrire un algorithme Python permettant d'échanger le premier élément avec le dernier élément d'une liste donnée.

```
bovins = ['Limousine', 'Aubrac', 'Charolaise', 'Simmental']

def echanger_premier_dernier(Liste_de_bovins) :
  bovins[0], bovins[-1] = bovins[-1], bovins[0]
  return Liste_de_bovins

print(echanger_premier_dernier(bovins))
```

**Exercice 7 :** Ecrire un algorithme Python qui renvoie la liste des diviseurs d'un entier donné.

```
def trouver_diviseurs(n) :
  diviseurs = []
  for i in range(1, n+1):
    if n % i == 0 :
      diviseurs.append(i)
  return diviseurs

print(trouver_diviseurs(12))
```

**Exercice 8 :** Ecrire un algorithme en python qui renvoie la moyenne des termes d'une liste de nombres.

```
entiers = [10, 20, 50, 90]

def calculer_moyenne(entiers) :
  return sum(entiers) / len(entiers)

print(calculer_moyenne(entiers))
```

**Exercice 9 :** Ecrire un algorithme en python sous forme de fonction qui prends en paramètre un couple (L, a) formé d'une liste L et d'un élément a et qui renvoie True si l'élément a est présent dans la liste L et False si non.

```
def est_present(L, a):
    if a in L:
        return True
    else:
        return False

print(est_present([10, 20, 30], 20))
print(est_present([10, 20, 30], 50))
```

**Exercice 10 :** Ecrire un algorithme en python sous forme de fonction qui prends en paramètre un couple (L, a) formé d'une liste L et d'un élément a et qui renvoie la position de l'élément a dans la liste L. La fonction doit renvoyer -1 si l'élément a n'est pas présent dans la liste.

```
caprins = ['Alpine', 'Saanen', 'Angora']

def chercher_position(L, a):
    for i in range(len(L)):
        if L[i] == a:
            return i
    return -1

print(chercher_position(caprins, 'Alpine'))
print(chercher_position(caprins, 'Aubrac'))
```
