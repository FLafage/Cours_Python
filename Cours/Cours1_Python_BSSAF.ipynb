Boucle if :

```python
if a==5:
  print("a est égal à 5")
else:
  print("a n'est pas égal à 5")
```

Création d'une fonction factorielle :

```python
def fact(n):
  if n==1:
    return 1
  else:
    return fact(n-1)*n

fact(4)
```

Boucle for :

```python
a = 1
for i in range(1,11):
  a = a*i
  print("factorielle de ", i, "égale", a)
```
